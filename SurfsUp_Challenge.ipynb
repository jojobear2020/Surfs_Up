```python
# Dependencies
import numpy as np
import pandas as pd

# Python SQL toolkit and Object Relational Mapper
import sqlalchemy
from sqlalchemy.ext.automap import automap_base
from sqlalchemy.orm import Session
from sqlalchemy import create_engine, func
import datetime

from matplotlib import style
style.use('fivethirtyeight')
import matplotlib.pyplot as plot
%matplotlib inline
```

```python
engine = create_engine("sqlite:///hawaii.sqlite")

# reflect an existing database into a new model
Base = automap_base()
# reflect the tables
Base.prepare(engine, reflect=True)

# Save references to each table
Measurement = Base.classes.measurement
Station = Base.classes.station
```

```python
# Create our session (link) from Python to the DB
session = Session(engine)
```

## D1: Determine the Summary Statistics for June

```python
# 1. Import the sqlalchemy extract function.
from sqlalchemy import extract

# 2. Write a query that filters the Measurement table to retrieve the temperatures for the month of June. 
june_tobs = session.query(Measurement.date, Measurement.tobs).filter(extract('month', Measurement.date) == 6).all()
print(june_tobs)
```

```python
#  3. Convert the June temperatures to a list.
type(june_tobs)
```

```python
# 5. Calculate and print out the summary statistics for the June temperature DataFrame.

# Save the query results as a Pandas DataFrame and set the index to the date column
df_jun = pd.DataFrame(june_tobs, columns=['date','temperature'])
df_jun.set_index(df_jun['date'], inplace=True)

# Sort the dataframe by date
df_jun = df_jun.sort_index()
print(df_jun.to_string(index=False))
```

```python
# use the describe() function to calculate the mean, minimum, maximum, standard deviation, and percentiles
df_jun.describe()
```

```python
#plot to see the most occuring temperatures in June
df_jun.plot.hist(bins=8, title = "Hawaii Temperature Trend in June (2010-2017)")
# save praph
plot.savefig("analysis/jun_temp_occurence.png")
```

```python
#plot June temp trend thru years
jun_graph = df_jun.plot(figsize = (16, 9),
                        sort_columns=True,
                        rot=45,
                        use_index=True,
                        legend=True,
                        grid=True,
                        color='b')
plot.ylabel('Temperature (Fahrenheit)', weight='bold')
plot.xlabel('Date', weight='bold')
plot.title("June Temperatures in Hawaii (2010-2017)", fontsize=20)
# Save the figure.
plot.savefig("analysis/jun_temp_trend.png")
plot.show()
```

## D2: Determine the Summary Statistics for December

```python
# 6. Write a query that filters the Measurement table to retrieve the temperatures for the month of December.
december_tobs = session.query(Measurement.date, Measurement.tobs).filter(extract('month', Measurement.date) == 12).all()
print(december_tobs)
```

```python
# 7. Convert the December temperatures to a list.
type(december_tobs)
```

```python
# 8. Create a DataFrame from the list of temperatures for the month of December. 
df_dec = pd.DataFrame(december_tobs, columns=['date','temperature'])
df_dec.set_index(df_dec['date'], inplace=True)

# Sort the dataframe by date
df_dec = df_dec.sort_index()
print(df_dec.to_string(index=False))
```

```python
# 9. Calculate and print out the summary statistics for the Decemeber temperature DataFrame.
df_dec.describe()
```

```python
# plot to see occurence trend
df_dec.plot.hist(bins=8, title = "Hawaii Temperature Trend in December (2010-2017)")
# save graph
plot.savefig("analysis/dec_temp_occurence.png")
```

```python
#plot December temp trend thru years
dec_graph = df_dec.plot(figsize = (16, 9),
                        sort_columns=True,
                        rot=45,
                        use_index=True,
                        legend=True,
                        grid=True,
                        color='b')
plot.ylabel('Temperature (Fahrenheit)', weight='bold')
plot.xlabel('Date', weight='bold')
plot.title("December Temperatures in Hawaii (2010-2017)", fontsize=20)
#save graph
plot.savefig("analysis/dec_temp_trend.png")
plot.show()
```
